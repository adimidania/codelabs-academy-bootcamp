3. Checking whether the Tensor is Diagonal or Triangular

```python
# the method count_nonzero counts the number of non-zero values in the array.
def isDiagonal(M):
    return (np.count_nonzero(M - np.diag(np.diagonal(M))) == 0)

isDiagonal(img_frame)
```

```python
def isTriangularL(M):
    return (np.count_nonzero(M - np.tril(M)) == 0)
isTriangularL(img_frame)
```

```python
def isTriangularU(M):
    return (np.count_nonzero(M - np.triu(M)) == 0)

isTriangularU(img_frame)
```
