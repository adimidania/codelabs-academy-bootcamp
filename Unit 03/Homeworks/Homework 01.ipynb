```python
# Checking how many columns we've got
print(len(df)) 
```

```python
# We add a new column that says whether a person has a food allergy or not
df.insert(len(df.columns), 'FOOD_ALLERGY', np.ones((len(df),), dtype=bool))
```

```python
df
```

```python
# If all the columns starting from SHELLFISH_ALG_START up to NUM_ASTHMARX are NAN
# This means that the person doesn't have any kind of allergy
# Otherwise, s/he has
for i in range(len(df)):
    df.at[i,'FOOD_ALG'] = df.iloc[i, 9:-1].any()
```

```python
# Dropping the allergy columns as well as the SUBJECT_ID
allergy_columns = df.columns[9:-1]
df.drop(allergy_columns, axis=1, inplace=True)
df.drop(["SUBJECT_ID"], axis=1, inplace=True)
```

```python
# Changing the type to integers (0,1)
df["FOOD_ALG"] = df["FOOD_ALG"].astype(int)
df["ATOPIC_MARCH_COHORT"] = df["ATOPIC_MARCH_COHORT"].astype(int)
```

```python
# Extracting only the meaningful elements from the objects.
# GENDER_FACTOR[1] will be either Female, Male or Unknown
for i in range(len(df)):
    df.at[i,"GENDER_FACTOR"] = int(df.at[i,"GENDER_FACTOR"][1])
    df.at[i,"RACE_FACTOR"] = int(df.at[i,"RACE_FACTOR"][1])
    df.at[i,"ETHNICITY_FACTOR"] = int(df.at[i,"ETHNICITY_FACTOR"][1])
    df.at[i,"PAYER_FACTOR"] = int(df.at[i,"PAYER_FACTOR"][1])
```

```python
df["GENDER_FACTOR"] = df["GENDER_FACTOR"].astype(int)
df["RACE_FACTOR"] = df["RACE_FACTOR"].astype(int)
df["ETHNICITY_FACTOR"] = df["ETHNICITY_FACTOR"].astype(int)
df["PAYER_FACTOR"] = df["PAYER_FACTOR"].astype(int)
```

## The logistic regression model

```python
from sklearn.model_selection import train_test_split

X = df.drop(['FOOD_ALG'],axis=1)
y = df['FOOD_ALG']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)
score = model.score(X_test,y_test)
```
