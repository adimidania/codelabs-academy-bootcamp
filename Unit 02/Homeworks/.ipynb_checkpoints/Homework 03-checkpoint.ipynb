```python
import pandas as pd
df = pd.read_csv("fish.csv")
```

```python
df
```

```python
# Checking null values
df.isna().sum()
```

```python
# Extracting the variables
x = df.iloc[:,2:]
y = df.iloc[:,1]
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x,y)
```

```python
r_squared = model.score(x,y)
print("The R squared is :",r_squared)
```
