### Spot the possible outliers in your dataset and get rid of them.

#### Calculating the z-scores

```python
df["views_zscore"] = (df.views - df.views.mean()) / df.views.std()
df["likes_zscore"] = (df.likes - df.likes.mean()) / df.likes.std()
df["dislikes_zscore"] = (df.dislikes - df.dislikes.mean()) / df.dislikes.std()
df["comment_count_zscore"] = (df.comment_count - df.comment_count.mean()) / df.comment_count.std()
```

```python
# Now I will keep only the rows that doesn't contain outliers
# I will remove all the outliers from all the numerical rows
df = df[abs(df["views_zscore"]) < 3]
df = df[abs(df["likes_zscore"]) < 3]
df = df[abs(df["dislikes_zscore"]) < 3]
df = df[abs(df["comment_count_zscore"]) < 3]
```

```python
# Dropping the columns zscore because we don't need them
df.drop(['views_zscore', 'likes_zscore', 'dislikes_zscore', 'comment_count_zscore'], axis = 1, inplace = True)
```

```python
# Extracting all the tags that we have
tags_collection = df.tags.unique()
tags = []
for i in range(0,len(tags_collection)):
    tags.extend(tags_collection[i].replace('"','').split("|"))
```

```python
# Getting only the unique tags
unique = []
for tag in tags:
    if tag not in unique:
        unique.append(tag)
```

### Top 10 Tags

```python
from collections import Counter

tags = df.tags.values
all_tags = []
for i in range(0,len(tags)):
    all_tags.extend(tags[i].replace('"','').split("|"))

Counter(all_tags).most_common(10)
```
